```python
#1
def safe_divide(a, b):
    try:
        result = a / b
        print("Result:", result)

    except ZeroDivisionError:
        print("Error: Cannot divide by zero.")

    except TypeError:
        print("Error: Invalid data type.")

    finally:
        print("Operation attempted.")


# Test cases
safe_divide(10, 2)
safe_divide(10, 0)
safe_divide("10", 2)
```

```python
#2
class InvalidAgeError(Exception):
    pass


def validate_age(age):
    if age < 0 or age > 120:
        raise InvalidAgeError("Age must be between 0 and 120.")
    print("Valid Age:", age)


ages = [25, -5, 150]

for age in ages:
    try:
        validate_age(age)
    except InvalidAgeError as e:
        print("Invalid Age Error:", e)
```

```python
#3
user_inputs = ['10', 'abc', '0', '5']

for value in user_inputs:
    try:
        num = int(value)
        result = 100 / num
        print("Result:", result)

    except ValueError:
        print("ValueError: Input is not a number.")

    except ZeroDivisionError:
        print("ZeroDivisionError: Cannot divide by zero.")

    except Exception as e:
        print("Other Error:", e)
```

```python
#4
import numpy as np

arr = np.arange(1, 21)

# a
arr_2d = arr.reshape(4, 5)
print(arr_2d)

# b
print("Row Sum:", arr_2d.sum(axis=1))
print("Column Sum:", arr_2d.sum(axis=0))

# c
print("Greater than 10:", arr[arr > 10])

# d
arr[arr % 2 == 0] = -1
print(arr)
```

```python
#5
import numpy as np

A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

# a
print("Addition:\n", A + B)
print("Subtraction:\n", A - B)
print("Multiplication:\n", A * B)

# b
print("Matrix Multiplication:\n", np.dot(A, B))

# c
print("Transpose of A:\n", A.T)

# d
print("Determinant:", np.linalg.det(A))
print("Inverse:\n", np.linalg.inv(A))
```

```python
#6
import numpy as np

np.random.seed(42)

arr = np.random.randint(1, 101, 15)

print("Array:", arr)

# a
print("Mean:", np.mean(arr))
print("Median:", np.median(arr))
print("Std Dev:", np.std(arr))
print("Min:", np.min(arr))
print("Max:", np.max(arr))

# b
print("Descending:", np.sort(arr)[::-1])

# c
indices = np.argsort(arr)[-3:]
print("Indices of 3 largest values:", indices)

# d
normalized = (arr - arr.min()) / (arr.max() - arr.min())
print("Normalized Array:", normalized)
```

```python
#7
import pandas as pd

data = {
    "Name": ["Aarav", "Priya", "Karan", "Meera", "Sahil"],
    "Age": [23, 21, 25, 22, 24],
    "Department": ["AI", "Data Science", "AI", "Web Dev", "Data Science"],
    "Score": [88, 92, 79, 95, 67]
}

df = pd.DataFrame(data)

# a
print(df.head(3))
print(df.dtypes)

# b
def grade(score):
    if score >= 90:
        return "A"
    elif score >= 75:
        return "B"
    else:
        return "C"

df["Grade"] = df["Score"].apply(grade)

# c
print(df.sort_values(by="Score", ascending=False))

# d
print(df[df["Department"] == "AI"])
```

```python
#8
# a
print(df.groupby("Department")["Score"].mean())

# b
print(df.loc[df["Score"].idxmax()])

# c
df.rename(columns={"Score": "Marks"}, inplace=True)
print(df)

# d
df2 = df.drop("Age", axis=1)
print(df2)
```

```python
#9
import pandas as pd
import numpy as np

data = {
    "Product": ["Pen", "Notebook", "Eraser", "Sharpener", "Ruler"],
    "Price": [10, np.nan, 5, 8, np.nan],
    "Stock": [100, 50, np.nan, 30, 20]
}

df = pd.DataFrame(data)

# a
print(df.isnull().sum())

# b
mean_price = df["Price"].mean()
df["Price"].fillna(mean_price, inplace=True)

# c
df = df.dropna(subset=["Stock"])

# d
df.to_csv("cleaned_products.csv", index=False)

print(df)
```
