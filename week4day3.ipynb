```python
try:
  f = open("/content/sample.txt")
  print(var1)
except FileNotFoundError:
  print("file does not exist")
except Exception as e:
  print(e)
else:
  print(f.read())
  f.close()
finally:
  print("run complete")
```

```python
try:
    f = open("/content/sample.txt")
    if f.name == 'sample.txt':
        raise Exception
except FileNotFoundError:
  print("file does not exist")
except Exception as e:
  print(e)
else:
  print(f.read())
  f.close()
finally:
  print("run complete")
```

```python
a = 10
b = 0
try:
  c = a/b
  print(c)
except ZeroDivisionError as e:
  print(e)
finally:
  print("run complete")
```
